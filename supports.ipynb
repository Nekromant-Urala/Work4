```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Настройка качественного отображения графиков в Colab
%config InlineBackend.figure_format = 'retina'
```

```python
x = - (0.15 * np.cos(np.pi/2) + 0.1) * np.sin(np.pi/4)
y = (0.15 * np.cos(np.pi/2) + 0.1) * np.cos(np.pi/4)
z = 1.0 + 0.15 * np.sin(np.pi/2)

print(x)
print(y)
print(z)
```

```python
x = - (0.15 * np.cos(np.pi/6) + 0.3) * np.sin(2*np.pi/3)
y = (0.15 * np.cos(np.pi/6) + 0.3) * np.cos(2*np.pi/3)
z = 1.0 + 0.15 * np.sin(np.pi/6)

print(x)
print(y)
print(z)
```

```python
t = np.linspace(0, 2, 200)

# Уравнения обобщенных координат (траектории звеньев)
# Звено 1: q1(t)
q1 = np.pi/4 + (25*np.pi/48)*t**3 - (25*np.pi/64)*t**4 + (5*np.pi/64)*t**5

# Звено 2: q2(t)
q2 = 0.1 + 0.25*t**3 - 0.1875*t**4 + 0.0375*t**5

# Звено 3: q3(t)
q3 = np.pi/2 - (5*np.pi/12)*t**3 + (5*np.pi/16)*t**4 - (np.pi/16)*t**5

x = - (0.15 * np.cos(q3) + q2) * np.sin(q1)
y = (0.15 * np.cos(q3) + q2) * np.cos(q1)
z = 1.0 + 0.15 * np.sin(q3)
```

```python
# 4. Визуализация
fig = plt.figure(figsize=(15, 10))
plt.subplots_adjust(hspace=0.3)

# График Звена 1
ax1 = fig.add_subplot(2, 2, 1)
ax1.plot(t, q1, 'tab:red', lw=2)
ax1.set_title('Траектория q1 (θ₁)')
ax1.set_ylabel('Радианы')
ax1.grid(True, linestyle='--')

# График Звена 2
ax2 = fig.add_subplot(2, 2, 2)
ax2.plot(t, q2, 'tab:green', lw=2)
ax2.set_title('Траектория q2 (d₂)')
ax2.set_ylabel('Метры')
ax2.grid(True, linestyle='--')

# График Звена 3
ax3 = fig.add_subplot(2, 2, 3)
ax3.plot(t, q3, 'tab:blue', lw=2)
ax3.set_title('Траектория q3 (θ₃)')
ax3.set_xlabel('Время (с)')
ax3.set_ylabel('Радианы')
ax3.grid(True, linestyle='--')

# 3D Траектория в пространстве
ax4 = fig.add_subplot(2, 2, 4, projection='3d')
ax4.plot(x, y, z, color='black', lw=3, label='Путь схвата')
ax4.scatter(x[0], y[0], z[0], color='red', s=50, label='Старт')
ax4.scatter(x[-1], y[-1], z[-1], color='blue', s=50, label='Финиш')
ax4.set_title('Движение схвата в XYZ')
ax4.set_xlabel('X')
ax4.set_ylabel('Y')
ax4.set_zlabel('Z')
ax4.legend()

plt.show()
```
